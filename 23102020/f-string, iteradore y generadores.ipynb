# f-string y el uso con print

```python
a = 10
b = 20
s = f"{a}, {b}"
print(s)
```

```python
a = 10
b = 20
s = f"{a=}, {b=}"
print(s)
```

```python
print(f"{id(10)=}")
print(f"{id(a)=}")
```

```python
print(f".{a:>80}.")
print(f".{a:<80}.")
print(f".{a:^80}.")
```

```python
print(f"en hexadecimal {a=:x}")
print(f"en octal {a=:o}")
print(f"en binario {a=:b}")
```

# iteradores

* Un **iterador** tiene previamente la lista de los elementos a generar
* Se utiliza <span style="color:red">iter()</span> para generar

```python
listaRange = list(range(10))
print(f"{listaRange=}")
listaRange = list(range(5,10))
print(f"{listaRange=}")
listaRange = list(range(10,5,-1))
print(f"{listaRange=}")
```

```python
lista = [1,4,7,10,2,5,8]
iterador = iter(lista)
for i in iterador:
    print(f"{i=}")
```

# Generadores

```python
def generador(n):
    contador = 0
    while contador < n:
        print(f"antes del yiedld {contador=}")
        yield contador
        contador += 1
        
```

```python
g = generador(5)
for i in g:
    print(f"{i=}")
    
```

```python
from random import randint
def generador2(max):
    n = 0
    while n < max:
        yield randint(1,10)
        n += 1
```

```python
g2 = generador2(5)
for i in g2:
    print(f"{i=}")
```

```python
from random import randint
def generador3(max):
    n = 0
    while n < max:
        valor = int(input("Dame un número entero"))
        yield valor
        n += 1
        
```

```python
g3 = generador3(5)
for i in g3:
    print(f"{i=}")
```

```python
def factorial(n):
    fac = 1
    for i in range(1,n+1):
        fac *= i
    return fac
```

```python
print(f"{factorial(5)}")
```

```python
print(f"{factorial(500)}")
```
